```python
import numpy as np
```

Создать массив со случайными значениями и максимальный элемент заменить на 0

```python
random_array = np.random.rand(3, 3)

print("Исходный массив:")
print(random_array)

max_value = np.max(random_array)
max_indices = np.where(random_array == max_value)
random_array[max_indices] = 0

print("\nМассив с замененным максимальным элементом на 0:")
print(random_array)
```

Создать матрицу 3x3 со значениями от 0 до 8

```python
matrix = np.arange(9).reshape(3, 3)

print(matrix)
```

Используя numpy создать и нормализовать матрицу.

```python
def normalize_matrix(matrix):
  return (matrix - np.min(matrix)) / (np.max(matrix) - np.min(matrix))
```

```python
print(normalize_matrix(matrix))
```
