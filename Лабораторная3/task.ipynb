```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
import math

def fix_skew(image, skew_angle):
    height, width = image.shape[:2]
    skewed_image = cv2.warpAffine(image, np.array([[1, skew_angle, -math.tan(skew_angle) * height], [0, 1, 0]]), (width, height))
    return skewed_image
```

```python
image = cv2.imread('image.jpg')

shifted_image = fix_skew(image, 0.5)

cv2_imshow(image)
cv2_imshow(shifted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
